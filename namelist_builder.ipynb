```python
def loads(s):
    try:
        return json.loads(s)
    except:
        print (s)
        return None
```

```python
x = list(map(loads,open('names.jsonl').read().split('\n')))
x = list(filter(lambda x: x is not None, x))
```

```python
names_df = pandas.DataFrame.from_records(x)
```

```python
t = names_df['count']
```

```python
new_count = []
for i in names_df['count']:
    try:
        new_count.append(int(i))
    except:
        new_count.append(int(i['$numberLong']))

names_df['count'] = new_count
```

```python
def trans_str(s):
    try:
        return transliterate.translit(s,reversed="True")
    except:
        return s
```

```python
names_df['trans_name'] = names_df.text.apply(trans_str)
```

```python
names_df.sort_values('count',ascending=False)
```

```python
names_df['text'] = names_df.text.str.lower()
```

```python
new_names = names_df.groupby('text').agg({'count':'sum','gender':'first'}).sort_values('count',ascending=False)

new_names = new_names.reset_index()
```

```python
new_names['trans_name']=new_names.text.apply(trans_str)
```

```python
new_names[new_names['count']>=5][['text','count']].to_csv('russian_names.csv',sep=',',index=False,header=False)

new_names[new_names['count']>=5][['trans_name','count']].to_csv('russian_trans_names.csv',sep=',',index=False,header=False)

new_names[new_names['count']>=5][['text']].to_csv('russian_names.txt',sep=',',index=False,header=False)

new_names[new_names['count']>=5][['trans_name']].to_csv('russian_trans_names.txt',sep=',',index=False,header=False)

```

```python
males = new_names[new_names.gender=='m']
males[males['count']>=5][['text','count']].to_csv('russian_male_names.csv',sep=',',index=False,header=False)

males[males['count']>=5][['trans_name','count']].to_csv('russian_male_trans_names.csv',sep=',',index=False,header=False)

males[males['count']>=5][['text']].to_csv('russian_male_names.txt',sep=',',index=False,header=False)

males[males['count']>=5][['trans_name']].to_csv('russian_male_trans_names.txt',sep=',',index=False,header=False)
```

```python
females = new_names[new_names.gender=='f']
females[females['count']>=5][['text','count']].to_csv('russian_female_names.csv',sep=',',index=False,header=False)

females[females['count']>=5][['trans_name','count']].to_csv('russian_female_trans_names.csv',sep=',',index=False,header=False)

females[females['count']>=5][['text']].to_csv('russian_female_names.txt',sep=',',index=False,header=False)

females[females['count']>=5][['trans_name']].to_csv('russian_female_trans_names.txt',sep=',',index=False,header=False)
```
